# Qcuit Lie-EQGNN Quark/Gluon Demo

This notebook is the pitch payload for the pip-first Qcuit pivot: a headless, importable workflow for HEP-style quantum graph models.

It uses a tiny toy quark/gluon-like dataset generated by `qcuit.hep`. The goal is to prove the software path: Lorentz-aware jet batching, LieEQGNN construction, and reproducible benchmarking. Replace the toy data with CMS/Open Data arrays for a real study.

## Install

From the package root during development:

```bash
pip install -e ".[hep,qml,dev]"
```

Use `quantum_backend="torch"` for fast CPU smoke tests. Switch to `quantum_backend="pennylane"` after installing the `qml` extra.

````python
from qcuit.hep import toy_quark_gluon_jets
from qcuit.models import LieEQGNN
from qcuit.benchmarks import benchmark_classifier, set_seed
````

````python
set_seed(11)
data = toy_quark_gluon_jets(num_jets=64, n_nodes=6, seed=11)

print("jets:", len(data))
print("nodes per jet:", data.n_nodes)
print("scalar features per node:", data.n_scalar)
print("labels:", data.labels.bincount().tolist())
````

````python
model = LieEQGNN(
    n_scalar=data.n_scalar,
    n_hidden=4,
    n_layers=2,
    quantum_blocks=["phi_e"],
    quantum_backend="torch",  # change to "pennylane" for real qnodes
)
````

````python
report = benchmark_classifier(
    model,
    data,
    epochs=10,
    batch_size=8,
    seed=11,
    device="cpu",
)

report.metrics
````

- `qcuit.hep` owns the Lorentz/jet graph data path.
- `qcuit.models.LieEQGNN` gives the lab a one-import model template.
- `qcuit.benchmarks` produces a headless result object that can be saved on an HPC cluster.
